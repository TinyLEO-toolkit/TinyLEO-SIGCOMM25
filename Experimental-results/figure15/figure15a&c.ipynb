# Figure 15a

```python
plt.rcParams['pdf.fonttype'] = 42
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111)
edge_colors=["w","w","w","w"]#["#4F86EC","#FBBD05","#CB4F40"]#['#3E5C70',"#E17257","#FAAF77"]
colors=['#3E5C70', '#1F77B4','#F9AE78','#E47158']
labels=["Starlink\n customers","Global Internet\nbackbone","Regional demands\n(Latin America)"]
location=[5,18,35]
bar_width=3
lw=1.5
bias=[-bar_width*3/2,-bar_width/2,bar_width/2,bar_width*3/2]
plt.grid(True, linestyle="--", linewidth=0.5,zorder=0)
LP_data=[1769,3485,1230]
plt.bar([item+bias[0]for item in location],[len(official_plt),len(backbone_plt),len(south_plt)],width=bar_width,color=colors[0],edgecolor=edge_colors[0],lw=lw,label="TinyLEO",zorder=3)
plt.bar([item+bias[1]for item in location],LP_data,width=bar_width,label="Gurobi",edgecolor=edge_colors[1],color=colors[1],zorder=3)
plt.bar([item+bias[2]for item in location],[6420,4884,2456],label="MegaReduce",width=bar_width,color=colors[2],edgecolor=edge_colors[1],lw=lw,zorder=3)
plt.bar([item+bias[3]for item in location],[6793]*3,lw=lw,width=bar_width,color=colors[3],edgecolor=edge_colors[3],label="Starlink",zorder=3)
print([len(official_plt),len(backbone_plt),len(south_plt)])
print(LP_data)
print([6420,4884,2456])
print([6793]*3)
plt.ylim(0,7000)
plt.xlim(-1.5,41)
plt.ylabel("Number of satellites",fontsize=18.5)
plt.xticks(location,labels,fontsize=15)
plt.yticks(fontsize=16)
plt.legend(fontsize=17,loc="lower right",ncol=1,bbox_to_anchor=(1.01,0.15))#loc="upper left",bbox_to_anchor=(-0.02,0.95)

plt.savefig("figure15a.pdf",bbox_inches="tight")
plt.savefig("figure15a.png",bbox_inches="tight")
```

# Figure 15c

```python
# official demand
official_plt=np.load("data/network_availability_official_demand.npy",allow_pickle=True)
backbone_plt=np.load("data/network_availability_backbone_demand.npy",allow_pickle=True)
south_plt=np.load("data/network_availability_Latin_America_demand.npy",allow_pickle=True)
```

```python
plt.rcParams['pdf.fonttype'] = 42
fig=plt.figure(figsize=(5,2.5))
colors=["#F57152","#FBB462","#AAE0E5","#6E78B9"]
colors = ['#2D9596','#74BCD5','#1F4999','#9F82CE']
colors=['#3E5C70','#E47158','#F9AE77',"#8D9B97"]
lables=["Starlink customers","Global Internet backbone","Regional demands\n(Latin America)"]
for idx,data in enumerate([official_plt,backbone_plt,south_plt]):
    data=np.append(data,[100]*6000)
    plt.plot(data,lw=3,color=colors[idx],label=lables[idx])
    
plt.legend(fontsize=15,loc='lower center', bbox_to_anchor=(0.5, 0))
plt.grid(True, linestyle="--", linewidth=0.5)
plt.xlim(0,7000)
plt.ylim(0,101)
plt.vlines(6793,0,100,color="#D84040",lw=2.5,ls="--")
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Network availability of"+"\n"+" $\geq$ 100Mbps per user(%)",fontsize=13)
plt.xlabel("Number of satellites",fontsize=18)
plt.savefig("figure15c.pdf",bbox_inches="tight")
plt.savefig("figure15c.png",bbox_inches="tight")
```
