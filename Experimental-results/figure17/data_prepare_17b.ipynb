```python
# routing updates command
import numpy as np
import networkx as nx
from tqdm import tqdm
import os
import time
from collections import defaultdict
import matplotlib.pyplot as plt

def get_satellite_connections(inter_topology, intra_topology):
    """Get all connections for each satellite"""
    connections = defaultdict(set)
    
    # Add inter-domain connections
    for (node1, node2), _ in inter_topology:
        connections[node1].add(node2)
        connections[node2].add(node1)
    
    # Add intra-domain connections
    for grid_connections in intra_topology.values():
        for node1, node2 in grid_connections:
            connections[node1].add(node2)
            connections[node2].add(node1)
    
    return connections

def analyze_topology_changes(prev_connections, curr_connections):
    """Analyze topology changes"""
    all_satellites = set(prev_connections.keys()) | set(curr_connections.keys())
    changed_satellites = set()
    
    # Detailed change statistics
    changes = {
        'lost_connections': defaultdict(set),  # Disconnected links
        'new_connections': defaultdict(set),   # New links
        'total_changed': 0,                    # Total number of satellites with changes
        'appeared': set(),                     # Newly appeared satellites
        'disappeared': set(),                  # Disappeared satellites
        'topology_changed': set()              # Satellites with topology structure changes
    }
    
    # Check each satellite
    for sat in all_satellites:
        prev_conns = prev_connections.get(sat, set())
        curr_conns = curr_connections.get(sat, set())
        
        # Check if satellite is newly appeared or disappeared
        if sat not in prev_connections:
            changes['appeared'].add(sat)
            changed_satellites.add(sat)
        elif sat not in curr_connections:
            changes['disappeared'].add(sat)
            changed_satellites.add(sat)
        else:
            # Check connection changes
            lost = prev_conns - curr_conns
            new = curr_conns - prev_conns
            
            if lost or new:
                changed_satellites.add(sat)
                changes['topology_changed'].add(sat)
                if lost:
                    changes['lost_connections'][sat] = lost
                if new:
                    changes['new_connections'][sat] = new
    
    changes['total_changed'] = len(changed_satellites)
    return changes

def process_topology_changes(all_inter_topology, all_intra_topology):
    """Process topology changes for all timestamps"""
    timestamps = sorted(all_inter_topology.keys())
    all_changes = []
    
    # Get connections for the first timestamp
    prev_connections = get_satellite_connections(
        all_inter_topology[timestamps[0]], 
        all_intra_topology[timestamps[0]]
    )
    
    # Initial statistics
    initial_stats = {
        'timestamp': timestamps[0],
        'total_satellites': len(prev_connections),
        'total_connections': sum(len(conns) for conns in prev_connections.values()) // 2
    }
    all_changes.append(initial_stats)
    
    # Process subsequent timestamps
    for t in tqdm(timestamps[1:], desc="Processing topology changes"):
        curr_connections = get_satellite_connections(
            all_inter_topology[t], 
            all_intra_topology[t]
        )
        
        # Analyze changes
        changes = analyze_topology_changes(prev_connections, curr_connections)
        
        # Record statistics
        stats = {
            'timestamp': t,
            'total_satellites': len(curr_connections),
            'total_connections': sum(len(conns) for conns in curr_connections.values()) // 2,
            'changed_satellites': changes['total_changed'],
            'appeared_satellites': len(changes['appeared']),
            'disappeared_satellites': len(changes['disappeared']),
            'topology_changed_satellites': len(changes['topology_changed']),
            'satellites_with_lost_connections': len(changes['lost_connections']),
            'satellites_with_new_connections': len(changes['new_connections']),
            'detailed_changes': changes
        }
        all_changes.append(stats)
        
        # Update previous timestamp state
        prev_connections = curr_connections
        
        # Print current timestamp statistics
        print(f"\nTimestamp {t}:")
        print(f"Total satellites: {stats['total_satellites']}")
        print(f"Total connections: {stats['total_connections']}")
        print(f"Changed satellites: {stats['changed_satellites']}")
        print(f"- Appeared: {stats['appeared_satellites']}")
        print(f"- Disappeared: {stats['disappeared_satellites']}")
        print(f"- Topology changed: {stats['topology_changed_satellites']}")
        print(f"- Lost connections: {stats['satellites_with_lost_connections']}")
        print(f"- New connections: {stats['satellites_with_new_connections']}")
    
    return all_changes

def plot_topology_changes(all_changes, output_dir):
    """Plot statistical charts of topology changes"""
    timestamps = [stat['timestamp'] for stat in all_changes[1:]]  # Skip first timestamp
    
    # 1. Number of changed satellites over time
    plt.figure(figsize=(15, 8))
    plt.plot(timestamps, [stat['changed_satellites'] for stat in all_changes[1:]], 
             label='Total Changed', color='black')
    plt.plot(timestamps, [stat['appeared_satellites'] for stat in all_changes[1:]], 
             label='Appeared', color='green')
    plt.plot(timestamps, [stat['disappeared_satellites'] for stat in all_changes[1:]], 
             label='Disappeared', color='red')
    plt.plot(timestamps, [stat['topology_changed_satellites'] for stat in all_changes[1:]], 
             label='Topology Changed', color='blue')
    
    plt.title('Satellite Changes Over Time')
    plt.xlabel('Timestamp')
    plt.ylabel('Number of Satellites')
    plt.legend()
    plt.grid(True)
    #plt.savefig(os.path.join(output_dir, 'satellite_changes_over_time_573_11_11_distancedt_global_max.png'))
    plt.show()
    
    # 2. Changes in total satellites and connections
    fig, ax1 = plt.subplots(figsize=(15, 8))
    
    ax1.set_xlabel('Timestamp')
    ax1.set_ylabel('Number of Satellites', color='tab:blue')
    ax1.plot(timestamps, [stat['total_satellites'] for stat in all_changes[1:]], 
             color='tab:blue', label='Total Satellites')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Number of Connections', color='tab:orange')
    ax2.plot(timestamps, [stat['total_connections'] for stat in all_changes[1:]], 
             color='tab:orange', label='Total Connections')
    ax2.tick_params(axis='y', labelcolor='tab:orange')
    
    plt.title('Network Scale Over Time')
    fig.tight_layout()
    #plt.savefig(os.path.join(output_dir, 'network_scale_over_time_573_11_11_distancedt_global_max.png'))
    plt.show()

def save_topology_changes(all_changes, output_dir):
    """Save topology change statistics"""
    # Save raw data
    np.save(os.path.join(output_dir, "topology_changes_573_11_11_distancedt_global_max.npy"), all_changes)

def print_topology_statistics(all_changes):
    """Print topology change statistics"""
    print("\nTopology Change Statistics:")
    
    # Overall statistics
    total_changes = sum(stat['changed_satellites'] for stat in all_changes[1:])
    avg_changes = np.mean([stat['changed_satellites'] for stat in all_changes[1:]])
    max_changes = max(stat['changed_satellites'] for stat in all_changes[1:])
    
    print(f"Total number of timestamps: {len(all_changes)}")
    print(f"Total changed satellites across all timestamps: {total_changes}")
    print(f"Average changed satellites per timestamp: {avg_changes:.2f}")
    print(f"Maximum changed satellites in a timestamp: {max_changes}")
    
    # Statistics by type
    total_appeared = sum(stat.get('appeared_satellites', 0) for stat in all_changes[1:])
    total_disappeared = sum(stat.get('disappeared_satellites', 0) for stat in all_changes[1:])
    total_topology = sum(stat.get('topology_changed_satellites', 0) for stat in all_changes[1:])
    
    print("\nChanges by Type:")
    print(f"- Total appeared satellites: {total_appeared}")
    print(f"- Total disappeared satellites: {total_disappeared}")
    print(f"- Total topology changed satellites: {total_topology}")
    
    # Network scale statistics
    avg_satellites = np.mean([stat['total_satellites'] for stat in all_changes])
    avg_connections = np.mean([stat['total_connections'] for stat in all_changes])
    
    print("\nNetwork Scale:")
    print(f"- Average number of satellites: {avg_satellites:.2f}")
    print(f"- Average number of connections: {avg_connections:.2f}")

if __name__ == "__main__":
    # Configuration parameters
    input_dir = "data"
    output_dir = "data"
    
    # Create output directory
    os.makedirs(output_dir, exist_ok=True)
    
    # Record start time
    start_time = time.time()
    
    # Load topology data
    print("Loading topology data...")
    all_inter_topology = np.load(os.path.join(input_dir, "all_inter_topology_573_11_11_distancedt_global_max.npy"), 
                                allow_pickle=True).item()
    all_intra_topology = np.load(os.path.join(input_dir, "all_intra_topology_573_11_11_distancedt_global_max.npy"), 
                                allow_pickle=True).item()
    
    # Analyze topology changes
    print("Analyzing topology changes...")
    all_changes = process_topology_changes(all_inter_topology, all_intra_topology)
    
    # Record end time
    end_time = time.time()
    print(f"\nTotal processing time: {end_time - start_time:.2f} seconds")
    
    # Generate visualizations
    print("Generating visualizations...")
    plot_topology_changes(all_changes, output_dir)
    
    # Save results
    print("Saving results...")
    save_topology_changes(all_changes, output_dir)
    
    # Print statistics
    print_topology_statistics(all_changes)
```
