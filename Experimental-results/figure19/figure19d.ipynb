```python
import numpy as np
without_failure = np.load('data/pcap/without_failure.npy')
with_failure = np.load('data/pcap/with_failure.npy')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Compute mean, min, and max for each condition
means = [np.mean(with_failure), np.mean(without_failure)]
mins = [np.min(with_failure), np.min(without_failure)]
maxs = [np.max(with_failure), np.max(without_failure)]

# Create figure
plt.figure(figsize=(3.3, 2))

# Compute error bars: [lower error, upper error]
yerr_lower = [mean - min_val for mean, min_val in zip(means, mins)]
yerr_upper = [max_val - mean for mean, max_val in zip(means, maxs)]
yerr = [yerr_lower, yerr_upper]

# Bar plot with error bars
conditions = ['w/ failure', 'w/o failure']
plt.bar(conditions, means, yerr=yerr, capsize=15,
        color=['#3D5C6F', '#E47159'], alpha=0.9)

# Axis labels
plt.ylabel('Pkt processing delay (ms)', fontsize=10)
plt.tick_params(axis='both', which='major', labelsize=11)

# Save figure
plt.savefig('figures/figure19d.pdf', dpi=300, bbox_inches='tight')
plt.savefig('figures/figure19d.png', dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
