```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
uniform_paths = np.load('data/uniform_paths.npy')
uniform_isls = np.load('data/uniform_isls.npy')
nonuniform_paths = np.load('data/nonuniform_paths.npy')
nonuniform_isls = np.load('data/nonuniform_isls.npy')

# Time axis (unit: minutes)
time = np.arange(len(uniform_paths)) * 5
print(len(uniform_paths))

# --- Plot: Number of broken ISLs ---
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(5, 3), height_ratios=[1, 1])

# Plot both datasets
for ax in [ax1, ax2]:
    ax.plot(time, uniform_isls, color='#3D5C6F', linewidth=2)
    ax.plot(time, nonuniform_isls, color='#E47159', linewidth=2)
    ax.grid(True, alpha=0.3)

# Y-axis limits and ticks
ax1.set_ylim(27200, 28400)
ax2.set_ylim(14400, 15000)
ax1.set_xlim(0, 500)
ax2.set_xlim(0, 500)
ax1.set_yticks([27400, 27800, 28200])
ax2.set_yticks([14400, 14800])

# Hide border between subplots
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False, top=False)
ax2.xaxis.tick_bottom()

# Add break marks
d = .015
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

# Tick formatting
for ax in [ax1, ax2]:
    ax.tick_params(axis='both', which='major', labelsize=14)

# Layout and export
plt.subplots_adjust(hspace=0.1)
plt.rcParams['pdf.fonttype'] = 42
plt.savefig('figure9a.png', dpi=300, bbox_inches='tight')
plt.savefig('figure9a.pdf', bbox_inches='tight')
plt.show()

# --- Plot: Shortest path length ---
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(5, 3), height_ratios=[1, 1])

# Plot both datasets
for ax in [ax1, ax2]:
    ax.plot(time, uniform_paths, color='#3D5C6F', linewidth=2)
    ax.plot(time, nonuniform_paths, color='#E47159', linewidth=2)
    ax.grid(True, alpha=0.3)

# Y-axis limits and ticks
ax1.set_ylim(352000, 360500)
ax2.set_ylim(229000, 233000)
ax1.set_xlim(0, 500)
ax2.set_xlim(0, 500)
ax1.set_yticks([353000, 356000, 359000])
ax2.set_yticks([229000, 232000])

# Hide border between subplots
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False, top=False)
ax2.xaxis.tick_bottom()

# Add break marks
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)

# Tick formatting
for ax in [ax1, ax2]:
    ax.tick_params(axis='both', which='major', labelsize=14)

# Layout and export
plt.subplots_adjust(hspace=0.1)
plt.savefig('figure9b.png', dpi=300, bbox_inches='tight')
plt.savefig('figure9b.pdf', bbox_inches='tight')
plt.show()
```
